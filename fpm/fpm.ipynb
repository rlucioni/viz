```python
%matplotlib inline

import json
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import operator
import math

from pattern import web
from fnmatch import fnmatch

# set some nicer defaults for matplotlib
from matplotlib import rcParams

import brewer2mpl

# brewer2mpl.get_map args: set name set type  number of colors
bmap = brewer2mpl.get_map('Set3', 'qualitative', 9)
# colors is a list of RGB triplets
colors = bmap.mpl_colors

# Set the random seed for consistency
np.random.seed(12)

# Change the default colors
rcParams['axes.color_cycle'] = colors

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['lines.linewidth'] = 2
rcParams['axes.grid'] = False
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'none'

rcParams['font.sans-serif'] = 'Helvetica Neue'
rcParams['font.family'] = 'sans-serif'
rcParams['font.weight'] = 200

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecessary plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
from HTMLParser import HTMLParser

class MLStripper(HTMLParser):
    def __init__(self):
        self.reset()
        self.fed = []
    def handle_data(self, d):
        self.fed.append(d)
    def get_data(self):
        return ''.join(self.fed)

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()
```

```python
html = requests.get("http://en.wikipedia.org/wiki/List_of_films_that_most_frequently_use_the_word_%22fuck%22").text
```

```python
films, years, fuck_counts, minutes, fpms = [], [], [], [], []

dom = web.Element(html)
table = dom.by_tag('table')[0]

for row in table.by_tag('tr')[1:]:
    films.append(strip_tags(row.by_tag('td')[0].content))
    years.append(row.by_tag('td')[1].content)
    fuck_counts.append(row.by_tag('td')[2].content)
    minutes.append(row.by_tag('td')[3].content)
    fpms.append(row.by_tag('td')[4].content)

df = pd.DataFrame({
    'film': films,
    'year': years,
    'fuck_count': fuck_counts,
    'minutes': minutes,
    'fpm': fpms,
})
```

Primary goal is to visualize a fucks-per-minute (fpm) ranking among films, and to graph the change over time of the average fpm per year.

```python
def hbar(title, item_count_dict, bar_color=colors[0], spacing=50, subplot=111):
    """
    Visualize sorted item and count data with a horizontal bar graph.
    
    Parameters
    -----------
    title: str
      the graph's title
        
    item_count_tuples: lst of tuples
      each tuple contains the word and its count
      
    bar_color: RGB triple (default: dark2_colors[0])
      the color of the bars
    
    spacing: int (default: 50)
      the spacing between the bars and their count labels
    """

    item_count_tuples = item_count_dict.items()
    get_count = operator.itemgetter(1)
    item_count_tuples = sorted(item_count_tuples, key=get_count, reverse=True)
    
    items, counts = [], []
    for item, count in item_count_tuples:
        items.append(item)
        counts.append(count)
    
    # plot
    pos = np.arange(len(items))
    
    if str(subplot)[2] == '1':
        plt.figure(figsize=(10, len(items)/3))
    
    plt.subplot(subplot)
    
    plt.title(title)
    plt.barh(pos, counts, color=bar_color)
    
    # add the numbers to the side of each bar
    for p, item, count in zip(pos, items, counts):
            plt.annotate(str(count), xy=(count + spacing, p + .5), va='center')
    
    # customize ticks
    ticks = plt.yticks(pos + .5, items)
    xt = plt.xticks()[0]
    plt.xticks(xt, [' '] * len(xt))
    
    # minimize chartjunk
    remove_border(left=False, bottom=False)
    plt.grid(axis = 'x', color ='white', linestyle='-')
    
    # set plot limits
    plt.ylim(pos.max() + 1, pos.min())
    
    plt.tight_layout()
```

```python
fpms_by_film = {}

for i in range(len(df)):
    fpms_by_film[str(df.ix[i]['film'])] = float(df.ix[i]['fpm'])

hbar("Uses of the word 'fuck' per minute, by film", fpms_by_film, spacing=.2)
```

```python
def sort_dict_to_lists(item_count_dict):
    item_count_tuples = item_count_dict.items()
    get_count = operator.itemgetter(1)
    item_count_tuples = sorted(item_count_tuples, key=get_count, reverse=True)
    
    items, counts = [], []
    for item, count in item_count_tuples:
        items.append(item)
        counts.append(count)
    
    return items, counts
```

```python
films, fpms = sort_dict_to_lists(fpms_by_film)

plt.figure(figsize=(25, 7))
plt.bar(range(len(fpms)), fpms, color=colors[0])
plt.grid(axis='y', color='white', linestyle='-', lw=1)
plt.title("Use of the word 'fuck' per minute (FPM), by film")
plt.xlabel('Film')
plt.ylabel('FPM')

plt.xticks(np.arange(len(fpms)), films, rotation=90)

remove_border(left=False)

# for x, y in zip(categories, counts_6002):
#     plt.annotate("%i" % y, (x, y+10), ha='center')
```

```python
fpms_by_year = {}

for year in df['year'].unique():
    fpms_by_year[int(year)] = []

for i in range(len(df)):
    fpms_by_year[int(df.ix[i]['year'])].append(float(df.ix[i]['fpm']))
```

```python
avg_fpms_by_year = {}

for year, fpms in fpms_by_year.iteritems():
    avg_fpms_by_year[year] = np.mean(fpms)
```

```python
plt.figure(figsize=(15, 7))
x = avg_fpms_by_year.keys()
y = avg_fpms_by_year.values()
plt.plot(x, y)
plt.title("Average FPMs by year")
plt.xlabel('Year')
plt.ylabel('FPM')
plt.ylim(0,3.5)
plt.xlim(1983, 2013)

params = np.polyfit(x, y, 1)
xp = np.linspace(min(x), max(x), 20)
yp = np.polyval(params, xp)
line = plt.plot(xp, yp, 'k', alpha=0.7)

# overplot error band
std_dev = np.std(y - np.polyval(params, x))
plt.fill_between(xp, yp - std_dev, yp + std_dev, color='k', alpha=0.075)

remove_border()
```
