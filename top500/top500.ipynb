```python
def get_top500_xml(year, month):
    "Year is an int, month is a string, either '06' or '11.'"
    if year not in range(1993, 2014):
        raise Exception("Invalid year.")
    if month != '06' and month != '11':
        raise Exception("Invalid month.")
    
    xml = requests.get("http://s.top500.org/static/lists/xml/TOP500_{}{}_all.xml".format(year, month)).text
    return xml
```

```python
def top500_to_df(xml):
    dom = web.Element(xml)
    
    ranks = [int(rank.content) for rank in dom.by_tag('top500:rank')]
    manufacturers = [manufacturer.content for manufacturer in dom.by_tag('top500:manufacturer')]
    r_maxs_mflops = [float(r_max.content)*1000 for r_max in dom.by_tag('top500:r-max')]
    r_maxs_tflops = [float(r_max.content)/1000 for r_max in dom.by_tag('top500:r-max')]
    countries = [country.content for country in dom.by_tag('top500:country')]
    years = [int(year.content) for year in dom.by_tag('top500:year')]

    df = pd.DataFrame({
        'rank':ranks,
        'manufacturer': manufacturers,
        'r-max mflops': r_maxs_mflops,
        'r-max tflops': r_maxs_tflops,
        'country': countries,
        'year': years
    })
    
    df = df.set_index('rank')
    
    return df
```

Primary goal is to visualize amount of computing power in teraFLOPS per country, and in megaFLOPS per capita by country.

```python
tflops_per_country = {}
mflops_per_capita = {}

for country in nov_2013['country'].unique():
    country_df = nov_2013[nov_2013['country'] == country]
    country_tflops = sum(country_df['r-max tflops'])
    country_per_capita_mflops = sum(country_df['r-max mflops'])/population[country]
    
    tflops_per_country[country] = round(country_tflops, 4)
    mflops_per_capita[country] = round(country_per_capita_mflops, 4)
    
```

```python
def hbar(title, item_count_dict, bar_color=colors[0], spacing=50, subplot=111):
    """
    Visualize sorted item and count data with a horizontal bar graph.
    
    Parameters
    -----------
    title: str
      the graph's title
        
    item_count_tuples: lst of tuples
      each tuple contains the word and its count
      
    bar_color: RGB triple (default: dark2_colors[0])
      the color of the bars
    
    spacing: int (default: 50)
      the spacing between the bars and their count labels
    """

    item_count_tuples = item_count_dict.items()
    get_count = operator.itemgetter(1)
    item_count_tuples = sorted(item_count_tuples, key=get_count, reverse=True)
    
    items, counts = [], []
    for item, count in item_count_tuples:
        items.append(item)
        counts.append(count)
    
    # plot
    pos = np.arange(len(items))
    
    if str(subplot)[2] == '1':
        plt.figure(figsize=(10, len(items)/3))
    
    plt.subplot(subplot)
    
    plt.title(title)
    plt.barh(pos, counts, color=bar_color)
    
    # add the numbers to the side of each bar
    for p, item, count in zip(pos, items, counts):
            plt.annotate(str(count), xy=(count + spacing, p + .5), va='center')
    
    # customize ticks
    ticks = plt.yticks(pos + .5, items)
    xt = plt.xticks()[0]
    plt.xticks(xt, [' '] * len(xt))
    
    # minimize chartjunk
    remove_border(left=False, bottom=False)
    plt.grid(axis = 'x', color ='white', linestyle='-')
    
    # set plot limits
    plt.ylim(pos.max() + 1, pos.min())
    
    plt.tight_layout()
```

```python
manufacturer_contribution_by_country = {}

for country in nov_2013['country'].unique():
    manufacturer_contribution_by_country[country] = {}
    country_df = nov_2013[nov_2013['country'] == country]
    for ix, row in country_df.iterrows():
        man = row['manufacturer']
        pwr = row['r-max tflops']
        
        if man in manufacturer_contribution_by_country[country]:
            manufacturer_contribution_by_country[country][man] += pwr
        else:
            manufacturer_contribution_by_country[country][man] = pwr
```

```python
country_tflop_tuples = tflops_per_country.items()
get_count = operator.itemgetter(1)
country_tflop_tuples = sorted(country_tflop_tuples, key=get_count, reverse=True)

countries, man_pwrs, total_pwrs = [], [], []
for country in [country for country, tflops in country_tflop_tuples]:
    countries.append(country)
    
    man_tflop_tuples = manufacturer_contribution_by_country[country].items()
    man_tflop_tuples = sorted(man_tflop_tuples, key=get_count, reverse=True)
    man_pwrs.append(man_tflop_tuples)
    
    total_pwrs.append(round(sum([pwr for man, pwr in man_tflop_tuples]), 2))

# plot

full_colors = brewer2mpl.get_map('Set3', 'qualitative', 12).mpl_colors

plt.figure(figsize=(10, len(countries)/2))

plt.title("TFLOPS Contributed by Manufacturer, per Country")

pos = np.arange(len(countries))

# build up each stacked bar
for num in pos:
    mans = man_pwrs[num]
    data_lst, color_lst = [], []
    for man, pwr in mans:
        data_lst.append(pwr)
        
        color_num = len(colors) - 1
        for rank, top_man_tup in enumerate(top_mans):
            if man == top_man_tup[0]:
                color_num = rank
                break
                
        if color_num == 1:
            color_lst.append(full_colors[11])
        else:
            color_lst.append(colors[color_num])

    left = 0
    for data, color in zip(data_lst, color_lst):
        plt.barh(num, data, color=color, left=left)
        left += data

# add the numbers to the side of each bar
spacing = 1500
for p, country, total_pwr in zip(pos, countries, total_pwrs):
    plt.annotate(str(total_pwr), xy=(total_pwr + spacing, p + .5), va='center')

# customize ticks
ticks = plt.yticks(pos + .5, countries)
xt = plt.xticks()[0]
plt.xticks(xt, [' '] * len(xt))

# minimize chartjunk
remove_border(left=False, bottom=False)
plt.grid(axis = 'x', color ='white', linestyle='-')

# set plot limits
plt.ylim(pos.max() + 1, pos.min())

plt.tight_layout()
```
