```python
%matplotlib inline

import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import os.path
import datetime

# set some nicer defaults for matplotlib
from matplotlib import rcParams

#these colors come from colorbrewer2.org. Each is an RGB triplet
dark2_colors = [(0.10588235294117647, 0.6196078431372549, 0.4666666666666667),
                (0.8509803921568627, 0.37254901960784315, 0.00784313725490196),
                (0.4588235294117647, 0.4392156862745098, 0.7019607843137254),
                (0.9058823529411765, 0.1607843137254902, 0.5411764705882353),
                (0.4, 0.6509803921568628, 0.11764705882352941),
                (0.9019607843137255, 0.6705882352941176, 0.00784313725490196),
                (0.6509803921568628, 0.4627450980392157, 0.11372549019607843),
                (0.4, 0.4, 0.4)]

rcParams['figure.figsize'] = (12, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.grid'] = False
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'none'
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = 'Helvetica'

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecessary plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
def record(exercise_name, date, reps, weight=None):
    if os.path.isfile('data/{}.json'.format(exercise_name)):
        f = open('data/{}.json'.format(exercise_name))
        data = json.loads(f.read())
        f.close
        
        d = {}
        d['date'] = date
        d['reps'] = reps
        
        if weight:
            d['weight'] = weight

        data.append(d)
        
        f = open('data/{}.json'.format(exercise_name), 'w')
        f.write(json.dumps(data, indent=2))
        f.close()
        
        print "Data recorded."
    else:
        f = open('data/{}.json'.format(exercise_name), 'w+')
        
        d = {}
        d['date'] = date
        d['reps'] = reps
        
        if weight:
            d['weight'] = weight
        
        f.write(json.dumps([d], indent=2))
        f.close()
        
        print "New exercise data file created. Data recorded."
```

```python
def retrieve(exercise_name):
    if os.path.isfile('data/{}.json'.format(exercise_name)):
        f = open('data/{}.json'.format(exercise_name))
        data = json.loads(f.read())
        return data
    else:
        raise Exception("No data recorded for {}.".format(exercise_name))
```

```python
def visualize(exercise_name, font_size=16):
    data = retrieve(exercise_name)
    
    dates, reps, weights = [], [], []
    
    for entry in data:
        dates.append(datetime.datetime.strptime(entry['date'], '%m-%d-%Y').date())
        reps.append(entry['reps'])
        
        try:
            weights.append(entry['weight'])
        except KeyError:
            pass
        
    plt.title("{}".format(exercise_name))
    plt.xlabel('Date')
    plt.xticks(rotation=300)
    plt.ylabel('Reps')
    plt.ylim(0,110)
    
    plt.plot(dates, reps, label='Reps', color=dark2_colors[0])
    
    if weights != []:
        def autolabel(rects):
            # attach some text labels
            for rect in rects:
                height = rect.get_height()
                plt.text(rect.get_x()+rect.get_width()/2., 0.05+height, '%d'%int(height),
                        ha='center', va='bottom')
    
        rects = plt.bar(dates, weights, align='center', label="Weight (lbs.)", color=dark2_colors[1])
            
        autolabel(rects)

    plt.legend(frameon=False, bbox_to_anchor=(1.25, 1.0), prop={'size':font_size})
    remove_border()
    plt.tight_layout()
```

```python
def small_multiples(exercise_lst, rows, cols, size):
    plt.figure(figsize=size)
    
    for n, exercise in enumerate(exercise_lst):
        plt.subplot(rows, cols, n+1)
        visualize(exercise, font_size=11)
    
    plt.subplots_adjust(wspace=0.5, hspace=None)
```
